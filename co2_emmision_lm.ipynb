```python
test.drop(["UvAerosolLayerHeight_aerosol_height", "UvAerosolLayerHeight_aerosol_pressure", "UvAerosolLayerHeight_aerosol_optical_depth", "UvAerosolLayerHeight_sensor_zenith_angle", "UvAerosolLayerHeight_sensor_azimuth_angle", "UvAerosolLayerHeight_solar_azimuth_angle", "UvAerosolLayerHeight_solar_zenith_angle"], axis=1, inplace=True)
```

```python
#converting data
from sklearn.preprocessing import StandardScaler
st_sclr = StandardScaler()
st_sclr.fit(train.drop("emission",axis=1))
train_ft = st_sclr.transform(train.drop("emission",axis=1))
```

```python
st_sclr.fit(test)
test_ft = st_sclr.transform(test)
```

```python
#transformed datasets
train_ft,test_ft
```

```python
#splitting
X_train = train_ft
y_train = train["emission"]
X_test = test_ft
```

```python
#importing knn model 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
```

```python
test_2 = pd.read_csv("test_co2.csv")
predictions = pd.DataFrame(predictions,columns=["emission"])
pred_2 = pd.concat([predictions,test_2["ID_LAT_LON_YEAR_WEEK"]],axis=1)
pred_2.set_index("ID_LAT_LON_YEAR_WEEK",inplace=True)
pred_2.to_csv("co2_predictions_2.csv")
```
